This purpose of this project is to extract a list of alumni and their respective companies from Entrepreneurship Lab Bio and Health Technology's directory page. With this data organized, it will be easier to visit alumni pages and see what their companies are up to.

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import urllib.request
import requests

html = 'https://www.elabnyc.com/participants/'
response = urllib.request.urlopen(html)
soup = bs(response, 'html.parser')
print(soup.prettify())
```

```python
# get list of names and links to their profiles
name_link = soup.findAll('h2', class_ = 'fl-post-title')
```

```python
name_company = soup.findAll('h2', class_ = 'fl-post-title')
# name = name_company.find.text.strip()
```

```python
# string of name
name = name_company[0].find().text.strip()
```

```python
# link of alum
link = name_company[0].a.get('href')
link
```
